```python
#creating bootstrap testing
def bootstrap_test(sample,iterations):
    means = np.empty(iterations)
    for i in range(iterations):
        new_sample = np.random.choice(sample,size=len(sample))
        n_mean = np.mean(new_sample)
        means[i]=n_mean
    mu = np.mean(means)
    std = np.std(means)
    return means, mu, std
```

```python
Temp = df['Temp']
```

```python
bs_means , mu, std = bootstrap_test(df['FireTotalAcres'],20000)
```

```python
_ = sns.distplot(bs_means)
sns.plt.show()
```
